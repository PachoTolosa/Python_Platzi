# 03_operations

```python
set_a = {"Col", "Mex", "Bol"}
set_b = {"Bol", "Per"}

# Union
# Union es la unión entre dos sets (set_a y set_b)
print("--------------------Union--------------------")
set_c = set_a.union(set_b)
print(set_c)
# ó
set_c = set_a | set_b
print(set_c)


# Intersection
# Intersection es la intersección entre dos sets (set_a y set_b) 
print("--------------------Intersection--------------------")
set_c = set_a.intersection(set_b)
print(set_c)
# ó
set_c = set_a & set_b
print(set_c)

# Difference
# Difference es la diferencia entre dos sets (set_a y set_b) 
print("--------------------Difference--------------------")
set_c = set_a.difference(set_b)
print(set_c)
# ó
set_c = set_a - set_b
print(set_c)

# Symmetric Difference
# Symmetric Difference es la diferencia entre dos sets, pero sin los elementos que se encuentran en ambos sets (set_a y set_b) 
print("--------------------Symmetric Difference--------------------")
set_c = set_a.symmetric_difference(set_b)
print(set_c)
# ó
set_c = set_a ^ set_b
print(set_c)

# Subset
# Subset es cuando un set está contenido en otro set (set_a está contenido en set_b) 
print("--------------------Subset--------------------")
set_c = set_a.issubset(set_b)
print(set_c)
# ó
set_c = set_a <= set_b
print(set_c)

# Superset
# Superset es lo contrario a subset (set_a contiene a set_b) 
print("--------------------Superset--------------------")
set_c = set_a.issuperset(set_b)
print(set_c)
# ó
set_c = set_a >= set_b
print(set_c)

# Disjoint
# Disjoint es cuando dos sets no tienen elementos en común (set_a y set_b no tienen elementos en común) 
print("--------------------Disjoint--------------------")
set_c = set_a.isdisjoint(set_b)
print(set_c)
```

# 11_game

```python
import random

def choose_options ():
    
    options = ('piedra', 'papel', 'tijera') 

    user_option = input('piedra, papel o tijera => ')
    user_option = user_option.lower()

    if not user_option in options:
        print('esa opcion no es valida')
    #continue
        return None, None

    computer_option = random.choice(options)

    print('User option =>', user_option)
    print('Computer option =>', computer_option)

    return user_option, computer_option


def check_rules (user_option, computer_option, user_wins, computer_wins):
    
    if user_option == computer_option:
        print('Empate!')
    elif user_option == 'piedra':
        if computer_option == 'tijera':
            print('piedra gana a tijera')
            print('user gano!')
            user_wins += 1
        else:
            print('Papel gana a piedra')
            print('computer gano!')
            computer_wins += 1
    elif user_option == 'papel':
        if computer_option == 'piedra':
            print('papel gana a piedra')
            print('user gano')
            user_wins += 1
        else:
            print('tijera gana a papel')
            print('computer gano!')
            computer_wins += 1
    elif user_option == 'tijera':
        if computer_option == 'papel':
            print('tijera gana a papel')
            print('user gano!')
            user_wins += 1
        else:
            print('piedra gana a tijera')
            print('computer gano!')
            computer_wins += 1
    return user_wins, computer_wins


def check_winner(user_wins, computer_wins):
    
    if computer_wins == 2:
        print('El ganador es la computadora')
        return True

    elif user_wins == 2:
        print('El ganador es el usuario')
        return True
    
    else:
        return False

def run_game():
    
    computer_wins = 0
    user_wins = 0
    rounds = 1

    while True:

        print('*' * 10)
        print('ROUND', rounds)
        print('*' * 10)

        print('computer_wins', computer_wins)
        print('user_wins', user_wins)
        
        rounds += 1

        user_option, computer_option = choose_options()
        user_wins, computer_wins = check_rules(user_option, computer_option, user_wins, computer_wins)
        
        if check_winner(user_wins, computer_wins):
            break
        
run_game()
```

```python
import csv

def read_csv(path):
    with open(path,"r") as csvfile:
        reader = csv.reader(csvfile, delimiter = ",")
        header = next(reader)
        print(header)
        data = []
        for row in reader:
            # print("***" * 5)
            iterable = list(zip(header, row))
            # print(iterable)
            country_dict = {key: value for key, value in iterable}
            # print(country_dict)
            data.append(country_dict)
        return data

if __name__ == "__main__":
    data = read_csv("./data.csv")
    print(data)
```

```python
import csv
import matplotlib.pyplot as plt

def read_csv(path):
    with open(path,"r") as csvfile:
        fileCSV = csv.DictReader(csvfile,delimiter=',')
        data = list(fileCSV)
    return data

data = read_csv("./data.csv")

data_by_continents = list(set(map(lambda continent: continent["Continent"], data)))
# print(data_by_continents)


print('Esta es la lista de continentes:')
print("")
continents = {}
text = ""
for i in range(len(data_by_continents)):
    continents[i+1] = data_by_continents[i]
    text += f'{continents[i+1]} = {i+1} \n'
    # print(f'{continents[i+1]} = {i+1}')
# print(continents)
print(text)


choosed_continent = int(input("Seleccione el numero del continente"))

filtered_data = list(filter(lambda item: item["Continent"] == continents[choosed_continent], data))
# print(filtered_data)

all_countries = list(map(lambda country: country['Country/Territory'], filtered_data))
all_population = list(map(lambda w_p_p: w_p_p['World Population Percentage'], filtered_data))

# all_countries = list(map(lambda country: country['Country/Territory'], data))
# all_population = list(map(lambda w_p_p: w_p_p['World Population Percentage'], data))


def generate_bar_chart(labels, values):
    plt.bar(labels, values)
    
def generate_pie_chart(labels, values):
    plt.figure(figsize=(20,10))
    plt.rcParams.update({"font.size": 10})
    plt.pie(values,labels=labels)
    
    
if __name__ == "__main__":
    # generate_bar_chart(labels, values)# Usar el primer subplot para el gráfico de barras
    generate_pie_chart(all_countries, all_population)
    plt.show()
```
